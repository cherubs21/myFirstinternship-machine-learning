```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Loan_Train.csv")
```

```python
df
```

```python
vals_to_replace = {'Male':'1', 'Female':'2'}
df['Gender'] = df['Gender'].map(vals_to_replace)
df
```

```python
vals_to_replace1 = {'Graduate':'1', 'Not Graduate':'2'}
df['Education'] = df['Education'].map(vals_to_replace1)
df
```

```python
vals_to_replace2 = {'No':'0', 'Yes':'1'}
df['Married'] = df['Married'].map(vals_to_replace2)
df
```

```python
vals_to_replace3 = {'No':'0', 'Yes':'1'}
df['Self_Employed'] = df['Self_Employed'].map(vals_to_replace3)
df
```

```python
df['Dependents'] = df['Dependents'].replace('3+','3')
```

```python
vals_to_replace4 = {'Urban':'0', 'Rural':'1','Semiurban':'2'}
df['Property_Area'] = df['Property_Area'].map(vals_to_replace4)
df
```

```python
vals_to_replace5 = {'N':'0', 'Y':'1'}
df['Loan_Status'] = df['Loan_Status'].map(vals_to_replace5)
df
```

```python
df=df.drop(["Loan_ID"], axis=1)
df
```

```python
mean_value=df['LoanAmount'].mean()
mean_value
```

```python
df['LoanAmount'].fillna(value=mean_value, inplace=True)
df
```

```python
mean_value=df['Credit_History'].mean()
mean_value
```

```python
df['Credit_History'].fillna(value=mean_value, inplace=True)
df
```

```python
X=df[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']]
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_blobs
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X_train, y_train)
```

```python
clf = DecisionTreeClassifier(max_depth=None, min_samples_split=2,
...     random_state=0)
scores = cross_val_score(clf, X, y, cv=2)
scores.mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```
